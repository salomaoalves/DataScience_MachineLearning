# **Normalization and Label Encoder**

```python
# do label encoder in Day_of_week feature
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
train.Day_of_week = label.fit_transform(train.Day_of_week)
```

```python
def norma(df, cols):
    from sklearn.preprocessing import Normalizer

    scaler = Normalizer()
    df[cols] = scaler.fit_transform(df[cols])
    
    return df
```

# **Feature Selection**

```python
from sklearn.tree import DecisionTreeRegressor

X = train.drop('lights', axis=1)
y = train[['lights']]

model = DecisionTreeRegressor()
model.fit(X, y)

# get importance
importance = model.feature_importances_
```

```python
# get just importance feature
features = X.columns
im = importance < 0.03
cols = list(features[im].values)
cols.append('lights')
train = train[cols]
```

# **Machine Learning**

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
def metricas(y_test, y_pred):
    import numpy as np
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import mean_absolute_error
    from sklearn.metrics import median_absolute_error

    rmse = (np.sqrt(mean_squared_error(y_test, y_pred)))
    mae = mean_absolute_error(y_test, y_pred)
    medae = median_absolute_error(y_test, y_pred)
    
    return {'rmse': rmse, 'mae': mae, 'medae': medae}
```

```python
def reg_methods(method, model, data, norm):
    from sklearn.model_selection import train_test_split
    from sklearn.model_selection import cross_val_score
    
    X = data.drop('lights', axis=1)
    y = data[['lights']]
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
    
    if norm:
        cols = list(X_train.columns[:-1].values)
        X_train = norma(X_train, cols)
        
    model.fit(X_train, y_train)
    p = model.predict(X_val)
    param = model.get_params()
    results = metricas(y_val, p)
    results_cross = cross_val_score(model, X, y, cv=5)
    
    print(method)
    print(results)
    print('Cross-Validation', np.mean(results_cross))
            
    return results
```

```python
outliers = 'None' # None, ZScore, IQR, ZScore&IQR
obs = 'With Normalization and With Feature Selection'
NORM = 1

overall_results = pd.read_excel('result.xlsx')
#overall_results = pd.DataFrame()

methods = ['linear_regression','lasso','random_forest',
           'gradient_boosting','xgboost']

for method in methods:

    if method == 'linear_regression':
        model = LinearRegression()

    elif method == 'lasso':
        model = Lasso()
        
    elif method == 'random_forest':
        reg = RandomForestRegressor()
            
    elif method == 'gradient_boosting':        
        model = GradientBoostingRegressor()
            
    elif method == 'xgboost':
        model = XGBRegressor(objective="reg:squarederror")
        
        
    results = reg_methods(method,model,train,NORM)
    
    results['Metodo'] = method
    results['ObservaoÃ§Ãµes'] = obs
    results['Outliers'] = outliers
    
    results_df = pd.DataFrame(results, index=[0])
    
    overall_results = pd.concat([overall_results,results_df])
    
overall_results.to_excel('result.xlsx',index=False)
```
